```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Desktop/iris-data.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
df.drop_duplicates(keep='first')
```

```python
df['class'].unique()
```

```python
df.loc[df['class']=='Iris setossa','class']='Iris-setosa'
df.loc[df['class']=='versicolor','class']='Iris-versicolor'
```

```python
df.loc[df['petal_width_cm'].isnull(),'petal_width_cm']=df['petal_width_cm'].median()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.boxplot()
```

```python

Q1 = np.quantile(df['sepal_length_cm'], 0.25)
Q3 = np.quantile(df['sepal_length_cm'], 0.75)
Q11 = np.quantile(df['sepal_width_cm'], 0.25)
Q33 = np.quantile(df['sepal_width_cm'], 0.75)
IQR=Q3-Q1
low_drop=Q1-1.5*IQR
low_dropp=Q11-1.5*IQR
high_drop=Q3+1.5*IQR
high_dropp=Q33+1.5*IQR
df.drop(df.loc[(df['sepal_length_cm']<low_drop) | (df['sepal_length_cm']>high_drop)].index,inplace=True)
df.drop(df.loc[(df['sepal_width_cm']<low_dropp) | (df['sepal_width_cm']>high_dropp)].index,inplace=True)
```
